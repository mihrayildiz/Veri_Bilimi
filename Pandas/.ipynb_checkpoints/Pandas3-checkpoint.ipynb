#### aggregate

```python
import pandas as pd
df = pd.DataFrame({'degisken1': [10,20,23,23,22,11],
                   'degisken2': [100,300,522,652,889,780],
                    'gruplar' : ["A", "B", "B", "A","B","C"]
})
```

```python
df.groupby("gruplar")
```

```python
import numpy as np
df.groupby("gruplar").aggregate([min,np.median,max])
```

```python
df.groupby("gruplar").aggregate({'degisken1': min, 'degisken2': max}) #değişken1 için min değerler, degisken2 için max değerler
```

### filter

```python
def  filter_func(x):
    return x["degisken1"].std()>9
```

```python
df.groupby("gruplar").filter(filter_func)
```

### transform

```python
df_a = df.iloc[:,0:2]
df_a.transform(lambda x : (x-x.mean())/x.std())
```

### apply

```python
df.apply(np.sum)
```

Değişkenler üzerinde hazır fonksiyonlardan faklı olarak işlemler yapmak istersek filter,apply,transform metodları tecih
edilir. 
Aggregate,filter,transform,apply metodları ileri birleştirme metodlarıdır.

### Pivot Tablolar

```python
titanic =sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
titanic.tail()
```

```python
titanic.groupby("sex")[["survived"]].mean()
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
titanic.pivot_table("survived", index = "sex", columns = "class" )
```
